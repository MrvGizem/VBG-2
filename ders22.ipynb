```python
# Veri Türlerini Manipüle Etme Örneği
import pandas as pd

# Örnek bir veri seti
data = {
    'Urun_ID': [1, 2, 3],               # Teknik: Integer / Ölçme: Nominal (Sadece etiket)
    'Memnuniyet': ['Az', 'Orta', 'Cok'], # Teknik: Object / Ölçme: Ordinal
    'Sicaklik': [22.5, 24.1, 19.8],     # Teknik: Float / Ölçme: Interval
    'Satis_Adedi': [100, 150, 80]       # Teknik: Integer / ML: Discrete / Ölçme: Ratio
}

df = pd.DataFrame(data)
```

```python

# Teknik veri türlerini kontrol edelim
print(df.dtypes)
```

```python

# 'Memnuniyet' sütununu makineye uygun 'Ordinal' bir yapıya sokalım (Haftaya hazırlık)
df['Memnuniyet'] = pd.Categorical(df['Memnuniyet'], categories=['Az', 'Orta', 'Cok'], ordered=True)
print(df['Memnuniyet'])
```
