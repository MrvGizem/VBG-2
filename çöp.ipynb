```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

X = np.array([1,2,3,4,5,20]).reshape(-1,1)
y = np.array([2,4,6,8,10,200]) 

#aykırı değerle regresyon
model = LinearRegression()
model.fit(X,y)

print(model.coef_) # 20 gibi uç bir değer yüzünden eğim çok büyük çıkıyor.
```

```python
plt.scatter(X,y)
plt.show()
```

```python
#Aykırı değeri kaldırarak regresyon
X2 = np.array([1,2,3,4,5]).reshape(-1,1)
y2 = np.array([2,4,6,8,10])

model2 = LinearRegression()
model2.fit(X2,y2)

print(model2.coef_)
```

```python
plt.scatter(X2,y2)
plt.show()
```
