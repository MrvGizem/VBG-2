```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "urun_kodu": ["101", "102", "103", "104"],         # sayısal görünüyor ama aslında kategori
    "memnuniyet": [1, 2, 3, 4],                        # ordinal
    "uretim_suresi": ["12.5", "13.0", None, "14.2"],   # string + missing
    "hata_sayisi": [0, 1, 0, 5],                       # integer
    "tarih": ["2026-02-01", "2026-02-02", "2026-02-03", "2026-02-04"],  # string tarih
    "gelir": ["10.000", "12.500", "9.750", "11.200"]   # TR biçimi (nokta binlik)
})

df
```

```python
df.iloc[0,0] #matris elemanını gösterir
```

```python
df.dtypes #DataFrame içindeki her sütunun veri tipini gösterir.
```

```python
df.info()
```

```python
df.select_dtypes(include="object").columns
```

```python
df["urun_kodu"] = df["urun_kodu"].astype("category")
df["urun_kodu"].dtype
```

```python
df.dtypes 
```

```python
from pandas.api.types import CategoricalDtype

ordinal_type = CategoricalDtype(categories=[1,2,3,4,5], ordered=True)
df["memnuniyet"] = df["memnuniyet"].astype(ordinal_type)

df["memnuniyet"].dtype, df["memnuniyet"].cat.ordered #ordinal değişkeni sıralı kategorik yapmak
```

```python
df["uretim_suresi"].mean() #string olan üretim süresi sütunu üzerinde ortalama alma işlemi hata verir. Önce sayısal hale getirmeliyiz.
```

```python
df["uretim_suresi"] = pd.to_numeric(df["uretim_suresi"], errors="coerce") 
df["uretim_suresi"].mean()
```

```python
df["gelir"].mean()
```

```python
df["gelir"] = df["gelir"].str.replace(".", "", regex=False)  # binlik noktayı kaldır
df["gelir"] = pd.to_numeric(df["gelir"], errors="coerce")
df["gelir"].mean()
```

```python
df["tarih"] = pd.to_datetime(df["tarih"], errors="coerce")
```
